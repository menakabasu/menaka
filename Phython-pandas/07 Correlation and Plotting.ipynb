Finding Relationships
---------------------

A great aspect of the Pandas module is the `corr()` method.

The `corr()` method calculates the relationship between each column in your data set.

The examples in this page uses a CSV file called: 'data.csv'.

[Download data.csv](https://www.w3schools.com/python/pandas/data.csv). or [Open data.csv](https://www.w3schools.com/python/pandas/data.csv.txt)

### Example: Show the relationship between the columns

Show the relationship between the columns:

```python
# 1. Import the pandas library (toolbox for data)
import pandas as pd

# 2. Load the data from a CSV file (open the data book)
df = pd.read_csv('data.csv')

# 3. Calculate correlations (explore relationships between topics)
print(df.corr())
```

**`corr()` method Explanation:**

- **Calculate correlations:** This line uses the `corr()` method to calculate correlations between all numerical columns in the DataFrame. Correlations measure how closely pairs of variables move together, like noticing patterns in how different topics in the book relate to each other.

**Output:**

- The `print(df.corr())` statement will display a correlation matrix, which is a table showing the correlation coefficients between each pair of numerical columns.
    - Values range from -1 to 1:
        - Positive values indicate a positive relationship (variables tend to increase or decrease together).
        - Negative values indicate a negative relationship (variables tend to move in opposite directions).
        - Values closer to 1 or -1 indicate stronger relationships.
        - Values closer to 0 indicate weaker relationships.

**Key points:**

- The code helps you explore potential relationships and patterns within the data, which can be essential for understanding trends, making predictions, and identifying key factors.
- Correlations can uncover hidden connections that might not be immediately apparent from simple observation.
- The correlation matrix provides a valuable overview of how different variables in the DataFrame interact with each other.

**Note:** The `corr()` method ignores "not numeric" columns.

### Result Explained

The Result of the `corr()` method is a table with a lot of numbers that represents how well the relationship is between two columns.

**The number varies from -1 to 1.**

1 means that there is a 1 to 1 relationship (a perfect correlation), and for this data set, each time a value went up in the first column, the other one went up as well.

0.9 is also a good relationship, and if you increase one value, the other will probably increase as well.

\-0.9 would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.

0.2 means NOT a good relationship, meaning that if one value goes up does not mean that the other will.

**What is a good correlation?** It depends on the use, but I think it is safe to say you have to have at least `0.6` (or `-0.6`) to call it a good correlation.

### Perfect Correlation:

We can see that "Duration" and "Duration" got the number `1.000000`, which makes sense, each column always has a perfect relationship with itself.

### Good Correlation:

"Duration" and "Calories" got a `0.922721` correlation, which is a very good correlation, and we can predict that the longer you work out, the more calories you burn, and the other way around: if you burned a lot of calories, you probably had a long work out.

### Bad Correlation:

"Duration" and "Maxpulse" got a `0.009403` correlation, which is a very bad correlation, meaning that we can not predict the max pulse by just looking at the duration of the work out, and vice versa.

Plotting
--------

Pandas uses the `plot()` method to create diagrams.

We can use Pyplot, a submodule of the Matplotlib library to visualize the diagram on the screen.

Read more about Matplotlib at [Matplotlib Tutorial](https://www.w3schools.com/python/matplotlib_intro.asp).

The examples in this page uses a CSV file called: 'data.csv'.

[Download data.csv](https://www.w3schools.com/python/pandas/data.csv) or [Open data.csv](https://www.w3schools.com/python/pandas/data.csv.txt)



###Example: Import pyplot from Matplotlib and visualize our DataFrame



```python
# 1. Gathering the tools for drawing and data:
import sys       # This helps with sending information out, like a messenger.
import matplotlib  # This is the main library for creating plots and charts.
matplotlib.use('Agg')  # This sets a drawing style that works in different environments.

# 2. Bringing in pandas for data:
import pandas as pd  # This library is like a toolbox for handling data in tables.
import matplotlib.pyplot as plt  # This gives us specific tools for plotting.

# 3. Loading the data from a CSV file:
df = pd.read_csv('data.csv')  # This opens a CSV file, like a spreadsheet, and stores it as a DataFrame.

# 4. Creating a plot from the data:
df.plot()  # This asks pandas to create a visual representation of the data in the DataFrame.

# 5. Saving the plot as an image:
plt.savefig(sys.stdout.buffer)  # This saves the plot as a file, like taking a snapshot of the chart.
sys.stdout.flush()  # This makes sure the saved image is sent out to be seen.
```

**Key points for pandas learners:**

- **Importing libraries:** It's essential to import the necessary tools before using them, just like gathering art supplies before painting. This is like gathering your art supplies. You bring in`sys` to help with output, `matplotlib` for drawing, and `pandas` for working with data.
- **Loading data:** `pd.read_csv()` is a common way to load data from CSV files into pandas DataFrames, which are like organized tables.
- **Creating plots:** `df.plot()` is a quick way to visualize the data in a DataFrame, helping you see patterns and relationships.
- **Saving plots:** `plt.savefig()` allows you to save the plot as an image file for sharing or further analysis.
- **Flushing output:** `sys.stdout.flush()` ensures the saved image is displayed or sent correctly, like delivering a finished painting.





````python
# 1. Gathering the tools for drawing and data:
import sys       # This helps with sending information out, like a messenger.
import matplotlib  # This is the main library for creating plots and charts.
matplotlib.use('Agg')  # This sets a drawing style that works in different environments.

# 2. Bringing in pandas for data:
import pandas as pd  # This library is like a toolbox for handling data in tables.
import matplotlib.pyplot as plt  # This gives us specific tools for plotting.

# 3. Loading the data from a CSV file:
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/WK10-Pandas Learner/Resources/data.csv')  # This opens a CSV file, like a spreadsheet, and stores it as a DataFrame.

# 4. Creating a plot from the data:
df.plot()  # This asks pandas to create a visual representation of the data in the DataFrame.

# 5. Saving the plot as an image:
plt.savefig('my_plot.png')  # Save the plot as a PNG image

from IPython.display import Image
plt.savefig('my_plot.png')
Image('my_plot.png')  # Display the plot inline
````

Scatter Plot
------------

Specify that you want a scatter plot with the `kind` argument:

`kind = 'scatter'`

A scatter plot needs an x- and a y-axis.

In the example below we will use "Duration" for the x-axis and "Calories" for the y-axis.

Include the x and y arguments like this:

`x = 'Duration', y = 'Calories'`

### Example: Scatter plot between Duration and Calories


```python
# 1. Import necessary libraries:
import pandas as pd  # Library for data manipulation and analysis
import matplotlib.pyplot as plt  # Library for creating visualizations

# 2. Load the data from a CSV file:
df = pd.read_csv('data.csv')  # Read the data into a DataFrame

# 3. Create a scatter plot:
df.plot(kind='scatter', x='Duration', y='Calories')  # Create a scatter plot
                                                       # - Use 'Duration' as the x-axis
                                                       # - Use 'Calories' as the y-axis

# 4. Display the plot:
plt.show()  # Show the generated plot
```

**Explanation:**

1. **Import Libraries:**
   - `pandas` is imported for working with data in a structured format (DataFrames).
   - `matplotlib.pyplot` is imported for creating visualizations.

2. **Load Data:**
   - `pd.read_csv('data.csv')` reads the data from a CSV file named 'data.csv' and stores it in a DataFrame called `df`.

3. **Create Scatter Plot:**
   - `df.plot(kind='scatter', x='Duration', y='Calories')` creates a scatter plot using the DataFrame's data:
     - `kind='scatter'` specifies a scatter plot as the desired plot type.
     - `x='Duration'` sets the 'Duration' column as the data for the x-axis.
     - `y='Calories'` sets the 'Calories' column as the data for the y-axis.

4. **Display Plot:**
   - `plt.show()` displays the created scatter plot, making it visible to the user.

**Purpose:**

- This code visualizes the relationship between 'Duration' and 'Calories' in the 'data.csv' file through a scatter plot. It helps explore potential patterns or correlations between these variables.

Histogram
---------

Use the `kind` argument to specify that you want a histogram:

`kind = 'hist'`

A histogram needs only one column.

A histogram shows us the frequency of each interval, e.g. how many workouts lasted between 50 and 60 minutes?

In the example below we will use the "Duration" column to create the histogram:

**Note:** The histogram tells us that there were over 100 workouts that lasted between 50 and 60 minutes.

###Example: Creating Histogram
```python
# **Colab-specific configuration for inline plotting:**
%matplotlib inline  # Display plots directly within the notebook

# Import necessary libraries:
import pandas as pd  # Library for data manipulation and analysis
import matplotlib.pyplot as plt  # Library for creating visualizations

# Load the data from the CSV file:
df = pd.read_csv('/content/drive/MyDrive/WK10-pandas/Resources/data.csv')

# Create a histogram of the "Duration" column:
df["Duration"].plot(kind='hist')  # Create a histogram

# Customize the plot (optional):
plt.title('Distribution of Durations')  # Add a title
plt.xlabel('Duration')  # Label the x-axis
plt.ylabel('Frequency')  # Label the y-axis

# Display the plot:
plt.show()
```

**Explanation:**

1. **Colab Backend Configuration:**
   - `%matplotlib inline` tells Colab to display plots within the notebook, eliminating the need for the first three lines.

2. **Import Libraries:**
   - Imports `pandas` for data handling and `matplotlib.pyplot` for plotting.

3. **Load Data:**
   - Reads the CSV data into a DataFrame.

4. **Create Histogram:**
   - `df["Duration"].plot(kind='hist')` creates a histogram of the "Duration" column.

5. **Customize Plot (Optional):**
   - Adds a title and axis labels for clarity.

6. **Display Plot:**
   - `plt.show()` renders the histogram.

**Key Points:**

- The `%matplotlib inline` line is crucial for Colab's plot display.
- Verify the file path to your 'data.csv' file.
- Customize the plot further using matplotlib's features.
- Explore other plot types based on your analysis needs.

````python
# **Colab-specific configuration for inline plotting:**
# Display plots directly within the notebook
%matplotlib inline

# Import necessary libraries:
import pandas as pd  # Library for data manipulation and analysis
import matplotlib.pyplot as plt  # Library for creating visualizations

# Load the data from the CSV file:
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/WK10-Pandas Learner/Resources/data.csv')

# Create a histogram of the "Duration" column:
df["Duration"].plot(kind='hist')  # Create a histogram

# Customize the plot (optional):
plt.title('Distribution of Durations')  # Add a title
plt.xlabel('Duration')  # Label the x-axis
plt.ylabel('Frequency')  # Label the y-axis

# Display the plot:
plt.show()
````
